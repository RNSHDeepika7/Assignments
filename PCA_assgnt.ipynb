```python
import pandas as pd
df=pd.read_csv("wine.csv")
df
```

```python

X=df.iloc[:,1:14]
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
SS=StandardScaler()
SS_X=SS.fit_transform(X)
SS_X=pd.DataFrame(SS_X)
SS_X.columns=list(X)
SS_X.head()
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(SS_X)
SS_X_pca = pca.transform(SS_X)
SS_X_pca.shape
```

```python
pc_data=pd.DataFrame(SS_X_pca)
```

```python
pc_data.iloc[:,0].var()
```

```python
pc_data.iloc[:,3].var()
```

```python
pca.explained_variance_ratio_
```

```python
pd.DataFrame(pca.explained_variance_ratio_)*100
```

```python
X_new=pc_data.iloc[:,0:3]
X_new
```

```python
#Cluster analysis
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=4,affinity='euclidean',linkage='complete')
Y=cluster.fit_predict(X_new)
Y=pd.DataFrame(Y)
Y.value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_new.iloc[:,0],X_new.iloc[:,1],c=cluster.labels_,cmap="rainbow")
```

```python
#kmeans
from sklearn.cluster import KMeans
cluster=KMeans(n_init=20,n_clusters=4)
Y=cluster.fit_predict(X_new)
Y=pd.DataFrame(Y)
Y.value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_new.iloc[:,0],X_new.iloc[:,1],c=cluster.labels_,cmap="rainbow")
```

```python
cluster.inertia_
```

```python
kinertias=[]
for i in range(1,11):
    from sklearn.cluster import KMeans
    cluster=KMeans(n_init=20,n_clusters=i)
    cluster.fit_predict(X)
    kinertias.append(cluster.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.scatter(range(1,11),kinertias)
plt.plot(range(1,11),kinertias,color='red')
plt.show()
```
