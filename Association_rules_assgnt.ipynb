```python
from apyori import apriori
```

```python
trans = []
for i in range(0, 11):
  trans.append([str(df.values[i,j]) for j in range(0, 15)])
```

```python
trans
```

```python
len(trans)
```

```python
rules = apriori(transactions = trans,
                min_support = 0.003,
                min_confidence = 0.2,
                min_lift = 3,
                min_length = 2, max_length = 2)
```

```python
rules
```

```python
results = list(rules)
results
```

```python
len(results)
```

```python
results[0]
```

```python
base_item=[]
add_item=[]
support = []
confidence = []
lift = []
```

```python
results[0][1] # support
results[0][2][0][0] # base item
results[0][2][0][1] # add item
results[0][2][0][2] # confidence
results[0][2][0][3] # lift
```

```python
for i in range(0,48):
    base_item.append(results[i][2][0][0])
    add_item.append(results[i][2][0][1])
    support.append(results[i][1])
    confidence.append(results[i][2][0][2])
    lift.append(results[i][2][0][3])
```

```python
pd.DataFrame(base_item)
pd.DataFrame(add_item)
pd.DataFrame(support)
pd.DataFrame(confidence)
pd.DataFrame(lift)
```

```python
newdata = pd.concat([pd.DataFrame(base_item),pd.DataFrame(add_item),pd.DataFrame(support),pd.DataFrame(confidence),pd.DataFrame(lift)], axis=1)
```

```python
newdata.columns = ['Base item', 'Add item', 'Support', 'Confidence', 'Lift']
```
