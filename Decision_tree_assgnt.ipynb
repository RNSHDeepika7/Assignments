```python
import pandas as pd
df=pd.read_csv("Company_Data.csv")
df
```

```python
df.head()
df.info()
df.describe()
df.isnull().sum()
df.hist()
df.boxplot()
df.corr()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for column in df.select_dtypes(include='object'):
  le.fit(df[column])
  df[column] = le.transform(df[column])
```

```python
df.head()
```

```python
Y=df["Sales"]
X=df.iloc[:,1:11]
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
SS=StandardScaler()
SS_X=SS.fit_transform(X)
SS_X=pd.DataFrame(SS_X)
SS_X.columns=list(X)
SS_X.head()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(SS_X,Y,test_size=0.30)
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor()
dt_model.fit(X_train, Y_train)
Y_pred_train=dt_model.predict(X_train)
Y_pred_test=dt_model.predict(X_test)

from sklearn.metrics import mean_squared_error
er1 = np.sqrt(mean_squared_error(Y_train,Y_pred_train))
er2 = np.sqrt(mean_squared_error(Y_test,Y_pred_test))
print("training accuracy score=",er1.round(2))
print("testing accuracy score=",er2.round(2))
```

```python
import pandas as pd
df=pd.read_csv("Fraud_check.csv")
df
```

```python
df.head()
df.info()
df.describe()
df.isnull().sum()
df.hist()
df.boxplot()
df.corr()
```

```python
import numpy as np
df['Taxable.Income'] = np.where(df['Taxable.Income'] <= 30000, "Risky", "Good")
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for column in df.select_dtypes(include='object'):
  le.fit(df[column])
  df[column] = le.transform(df[column])
df.head()
```

```python
Y=df["Taxable.Income"]
X=df.drop("Taxable.Income", axis=1)
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
SS=StandardScaler()
SS_X=SS.fit_transform(X)
SS_X=pd.DataFrame(SS_X)
SS_X.columns=list(X)
SS_X.head()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(SS_X,Y,test_size=0.30)
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor()
dt_model.fit(X_train, Y_train)
Y_pred_train=dt_model.predict(X_train)
Y_pred_test=dt_model.predict(X_test)

from sklearn.metrics import mean_squared_error
er1 = np.sqrt(mean_squared_error(Y_train,Y_pred_train))
er2 = np.sqrt(mean_squared_error(Y_test,Y_pred_test))
print("training accuracy score=",er1.round(2))
print("testing accuracy score=",er2.round(2))
```
