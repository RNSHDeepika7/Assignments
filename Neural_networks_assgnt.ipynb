```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
!pip install --upgrade tensorflow
```

```python
from tensorflow.keras.metrics import RootMeanSquaredError
```

```python
rmse = RootMeanSquaredError()
```

```python
ann.compile(optimizer='adam', loss='mean_squared_error',metrics=[rmse])
```

```python
history = ann.fit(X_SS, y, validation_split=0.30, batch_size=100, epochs=50)
```

```python
ann.evaluate(X_SS, y)
```

```python
# summarize history for RMSE
import matplotlib.pyplot as plt
plt.plot(history.history['root_mean_squared_error'])
plt.plot(history.history['val_root_mean_squared_error'])
plt.title('model RMSE')
plt.ylabel('RMSE')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
import pandas as pd
df=pd.read_csv("forestfires.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
df["size_category"] = LE.fit_transform(df["size_category"])
df["month"] = LE.fit_transform(df["month"])
df["day"] = LE.fit_transform(df["day"])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Check for missing values
df.isnull().sum()

# Distribution of size category
sns.countplot(x='size_category', data=df)
plt.show()

# Distribution of FFMC
sns.countplot(x='FFMC', data=df)
plt.show()

# Distribution of DMC
sns.countplot(x='DMC', data=df)
plt.show()

# Distribution of DC
sns.countplot(x='DC', data=df)
plt.show()

# Distribution of age
sns.distplot(df['size_category'])
plt.show()

# Correlation matrix
corr = df.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

```python
X = df.iloc[:, 0:-1]
y = df.iloc[:, -1]
```

```python
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_SS = sc.fit_transform(X)
```

```python
y.head()
```

```python
import tensorflow as tf

tf.__version__
```

```python
ann = tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units=9, activation='relu'))
ann.add(tf.keras.layers.Dense(units=14, activation='relu'))
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
history = ann.fit(X_SS, y,validation_split=0.30, batch_size = 100, epochs = 100)
```

```python
scores = ann.evaluate(X_SS, y)
print("%s: %.2f%%" % (ann.metrics_names[1], scores[1]*100))
```

```python
history.history.keys()
```

```python
# summarize history for accuracy
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
